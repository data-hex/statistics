```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('median.csv', decimal=",")
df = df.rename(columns={"Unnamed: 0": "Область/Край/Республика", '20191)': "2019", '20201)': '2020', "20211)": "2021"})
df = df.set_index('Область/Край/Республика')
df = df.dropna()
df
```

```python
df = df.drop(labels = ['Республика Крым'],axis = 0) #удаляем Крым
df['2013'] = [float(x.replace(',', '.')) for x in df['2013']]
df['2014'] = [float(x.replace(',', '.')) for x in df['2014']]
#df = df.drop(columns = ['2013', '2014'], axis=1)
df
```

```python
# Сброс ограничений на количество выводимых рядов
pd.set_option('display.max_rows', None)
 
# Сброс ограничений на число столбцов
pd.set_option('display.max_columns', None)
 
# Сброс ограничений на количество символов в записи
pd.set_option('display.max_colwidth', None)
```

```python
sort_income = df.sort_values(by='2021')
```

```python
#отсортированные по 2021 году данные в порядке возрастания
sort_income
```

```python
sort_oblast = df.sort_index(ascending=True)
sort_oblast
```

```python
#ИЗМЕНЕНИЕ ПО ГОДАМ
fig = plt.figure(figsize=(6, 200))
total_axes = df.shape[0]
subplt = 1 # номер начального axes
for i, row in df.iterrows():
    ax = fig.add_subplot(total_axes, 1, subplt) # добавляем axes для каждого графика
    ax.set_title(i)
    plt.plot(df.loc[i]) #отрисовываем график
    subplt+=1
    
plt.tight_layout()
plt.show()
```
